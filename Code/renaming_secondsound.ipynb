```python
import pandas as pd
import os
```

```python
# Load the Excel file (update with your file path)
filename = "/Users/jorgegarciaponce/Desktop/Classes/Senior Spring/Physics 191/Superfluid Helium/Data/second_sound_filenames.xlsx"

# Load only the first three columns (assuming they are columns A, B, C)
df = pd.read_excel(filename, usecols="A:C", names=["Distance", "Pressure", "Filename"])

# Normalize filenames to lowercase for easier matching later
df['Filename'] = df['Filename'].str.lower()

print(df)
```

```python
# Define the directory where the files are
target_directory = "/Users/jorgegarciaponce/Desktop/Classes/Senior Spring/Physics 191/Superfluid Helium/Data/SecondSound"  # Adjust this path

# Supported file extensions
extensions = ['.png', '.set', '.csv']

# Loop through each file in the directory
for file in os.listdir(target_directory):
    # Check if the file (without extension) matches any entry in the Filename column
    name, ext = os.path.splitext(file)
    
    if ext in extensions:
        # Check if the file "name" (without extension) matches the Filename column
        match = df[df['Filename'] == name]
        
        if not match.empty:
            # Get corresponding pressure and distance
            pressure = match.iloc[0]['Pressure']
            distance = match.iloc[0]['Distance']
            
            # Create new filename
            new_name = f"{pressure}_{distance}{ext}"
            old_path = os.path.join(target_directory, file)
            new_path = os.path.join(target_directory, new_name)
            
            # Rename the file
            os.rename(old_path, new_path)
            print(f"Renamed: {file} -> {new_name}")
        else:
            print(f"No match for {file}, skipping.")

print("Renaming complete.")
```
